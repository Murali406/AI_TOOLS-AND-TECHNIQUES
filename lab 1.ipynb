```python
from matplotlib import pyplot as plt
```

```python
import numpy as np
```

```python
from matplotlib import style
```

```python
x=np.array([1,2,3])
y=np.array([5,7,4])
plt.plot(x,y,label="line")
plt.title("sample linear graph")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.legend()
plt.show()
```

```python
print(plt.style.available)
```

```python
x1=np.array([5,8,10])
y1=np.array([12,6,6])
x2=np.array([6,9,10])
y2=np.array([6,15,7])
plt.plot(x1,y1,label="line one",color='m',linewidth=6)
plt.plot(x2,y2,label="line two",color="g",linewidth=6)
plt.title("sample linear graph")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.legend()
plt.grid(True,color='k')
style.use("dark_background")
plt.show()
```

```python
x1=np.array([5,8,10])
y1=np.array([12,6,6])
x2=np.array([6,9,10])
y2=np.array([6,15,7])
plt.bar(x1,y1,label="bar one",color='m')
plt.bar(x2,y2,label="bar two",color='c')
plt.legend()
plt.xlabel("bar number")
plt.ylabel("bar values")
plt.title("sample bar graph")
plt.show()
```

```python
x=[10,15,20,21,22,45,56,67,23,12,80]
y=[0,10,20,30,40,50,60,70,90,100]  #increasing order

plt.hist(x,y,histtype="bar",label="hist",rwidth=0.5) #syntax important

plt.xlabel("x hist")
plt.ylabel("y hist")

plt.title("histogram")
plt.legend()
plt.show()
```

```python
x1=np.array([5,8,10])
y1=np.array([12,6,6])
plt.scatter(x1,y1,label="scat",color='r')
plt.xlabel("x scat")
plt.ylabel("y scat")
plt.title("scatterplot")
plt.legend()
plt.show()
```

```python
a=np.array([1,2,3,4,5])
b=np.array([2,3,4,3,2])
c=np.array([7,6,5,8,9])
d=np.array([8,7,5,4,10])
plt.plot([],[],color='g',label="green",linewidth=5)
plt.plot([],[],color='m',label='pink',linewidth=5)
plt.plot([],[],color='b',label='blue',linewidth=5)
plt.plot([],[],color='r',label='red',linewidth=5)
plt.stackplot(a,b,c,d,colors=['g','m','b','r'])   #syntax

plt.xlabel("x stack")
plt.ylabel("y stack")
plt.title("stcak plot")
plt.legend()
plt.show()
```

```python
a=np.array([10,3,4,12])
b=np.array(['green','blue','magenta','red'])
colors=['g','b','m','r']
plt.pie(a,labels=b,colors=colors,startangle=90,shadow=True)   #syntax
plt.title("piechart")
plt.show()
```
